```python
df["Age"].mean()
```

```python
df["Age"].fillna(df["Age"].mean(),inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["Sex"]
```

```python
df["Sex"] = df["Sex"].replace("male","Male")
df["Sex"] = df["Sex"].replace("female","Female")
```

```python
df
```

```python
df["Age"] = df["Age"].astype('int')
df
```

## Data Transformation 

```python
df["Family_count"] = df["SibSp"]+df["Parch"]
df
```

```python
df.drop(["SibSp"],axis=1,inplace=True)
df.drop(["Parch"],axis = 1, inplace = True)
df
```

```python
#group by
Pclass = df.groupby("Pclass")
Pclass_count = Pclass["Pclass"].count()
Pclass_count
```

```python
survived_not = df.groupby(["Survived","Sex"])["PassengerId"].count()
survived_not
```

## Data Visualization

```python
#boxplot
plt.figure(figsize =(8,6))
sns.barplot(x="Sex",y="Survived",data =df,color="darkblue",edgecolor="black")
plt.title("Gender vs Survived")
plt.xlabel("Gender")
plt.ylabel("Survied")
plt.show()
```

```python
#histogram
#age distribution
plt.figure(figsize=(8,6))
sns.histplot(df["Age"],bins=20,color="teal",kde=True)
```

```python
#scatter plot
sns.scatterplot(x="Age",y="Fare",data=df)
```

```python
sns.scatterplot(x="Age",y="Fare",data=df,hue=df['Pclass'])
```

```python
plt.figure(figsize=(8,3))
sns.scatterplot(x="Age",y="Fare",data=df,hue=df['Pclass'],palette ="viridis")
```
